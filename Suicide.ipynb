# Suicide Rates in Germany

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline 
plt.style.use('seaborn')

suicide=pd.read_csv('Suicide.csv', index_col = 0)
```

```python
suicide.describe()
```

```python
suicide.isnull().any()
```

```python
# Avarage age of suicide for each gender
suicide.groupby('sex').age.mean()
```

```python
#See all max values: max suicide freq, method, age
suicide.groupby('sex').max()
```

```python
# Visualization via boxplots, to see age and sex suicide relations
plt.figure(figsize=(10,6), dpi= 80)
ax = sns.boxplot(x=suicide.sex, y=suicide.age)
plt.title('Suicide Rates in Germany by sex/age', fontsize=15)
ax.set_xticklabels(['male', 'female'], rotation=45, fontsize=15)
plt.xlabel('Pohlavi')
plt.ylabel('VÄ›k')
plt.show()
```

```python
# Visualization via catplot, to see genders' preferences of different suicide methods
g=sns.catplot(x='method2', y='Freq', hue='sex', data=suicide, palette="dark", height=9, aspect=9/6, s=15, alpha=0.7)
plt.title('Suicide Rates in Germany, depends on method and sex', fontdict=dict(fontsize=20))
plt.xlabel('Method', fontdict=dict(fontsize=20))
plt.ylabel('Frequency', fontdict=dict(fontsize=20))
g._legend.set_title('Sex')
new_labels = ['Male', 'Female']
g.set_xticklabels(['poison', 'gas', 'hang', 'drown', 'gun','knife','jump', 'other'], fontdict=dict(fontsize=20))
plt.show()
```

```python
for age in suicide.age.unique():
    print('age',age,suicide[suicide.age == age].method.describe())
fig, ax = plt.subplots(figsize=(10,6), dpi= 80)
sns.boxplot(x='method', y='age', data=suicide)
```
